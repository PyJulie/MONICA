```python
import torch
import matplotlib.pyplot as plt
model = torch.load('../outputs/isic/500_ERM_224_resnet50_True_256_1_50/best.pt')

print(model.fc.weight.shape)
mean_values = model.fc.weight.mean(dim=1).cpu().detach().numpy()
min_value = mean_values.min()
max_value = mean_values.max()

# 归一化到 0-1 之间
normalized_values = (mean_values - min_value) / (max_value - min_value)
values_head = normalized_values[:2].mean()
values_medium = normalized_values[2:5].mean()
values_tail = normalized_values[5:8].mean()
# 生成条形图
plt.bar(range(1, 4), [values_head, values_medium, values_tail])
print([values_head, values_medium, values_tail])
plt.xlabel('Row')
plt.ylabel('Mean Value')
plt.title('Mean Value of Each Row in the Tensor')
plt.show()
```
